```python
from transformers import pipeline
pipe = pipeline("fill-mask", model="bert-base-uncased")
```

```python
output = pipe("Paris is the [MASK] of France")
output = pipe("Paris is the capital of [MASK].")
for element in output:
    print(f"Token {element['token_str']}:\t{element['score']:.3f}%")
```

```python
pipe = pipeline("fill-mask", model="xlm-roberta-base")
output = pipe("Paris is the capital of <mask>.")
for element in output:
    print(f"Token {element['token_str']}:\t{element['score']:.3f}%")
```

```python
research_finding = "Smoking causes lung cancer."
prompt = "The relationship described in this research finding is <mask>."
output = pipe(research_finding + prompt, targets=["causal", "association"])
for element in output:
    print(f"Token {element['token_str']}:\t{element['score']:.3f}%")
```
